```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Working\ML_Project\telecom_feature_engineered.csv")

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

#Traffic Trend Chart

```python
plt.figure(figsize=(12,6))

for cell in df['Cell_ID'].unique()[:5]:
    temp = df[df['Cell_ID'] == cell]
    plt.plot(temp['Date'], temp['DL_Volume_GB'], label=cell)

plt.title("DL Traffic Trend per Cell")
plt.xlabel("Date")
plt.ylabel("DL Volume (GB)")
plt.legend()

plt.show()
```

```python
#PRB Utilization Distribution
plt.figure(figsize=(7,5))

sns.histplot(df['PRB_Util'], bins=30)

plt.title("PRB Utilization Distribution")
plt.xlabel("PRB Utilization (%)")

plt.show()
```

```python
#Active  User vs Traffic

plt.figure(figsize=(7,5))

sns.scatterplot(x=df['Active_Users'], y=df['DL_Volume_GB'])

plt.title("Active Users vs Traffic")
plt.xlabel("Active Users")
plt.ylabel("DL Volume (GB)")

plt.show()
```

```python
#Correlation Heatmap

plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")

plt.show()
```
